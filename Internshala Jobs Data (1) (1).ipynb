```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# The Webpage URL
url = "https://internshala.com/jobs/"
```

```python
# HTTP Request
webpage = requests.get(url)
```

```python
webpage
```

```python
# webpage.content 
```

```python
#Using BreautifulSoup
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
# print(soup.prettify())
```

```python
# Scraping Job Titles
Job_Title = []
job_title = soup.find_all("h3", class_ = "job-internship-name")
for i in job_title:
    # print(i.text)
    Job_Title.append(i.text.replace('\n', ' '))
```

```python
# Scraping Company Names
Company_Name = []
company_name = soup.find_all("p", class_ = "company-name")
for i in company_name:
    # print(i.text)
    Company_Name.append(i.text.replace('\n', ' '))
```

```python
# Scraping Experience Required for Jobs
Experience_Required = []
exp_required = soup.find_all("div", class_ = "item_body desktop-text")
for i in exp_required:
    # print(i.text)
    Experience_Required.append(i.text)
```

```python
# Scraping Salary 
Salary_Per_Annum = []
job = soup.find_all("span", class_="desktop")
for i in job:
    # print(i.text)
    Salary_Per_Annum.append(i.text.replace('\n', ' '))
```

```python
# Creating a Data Frame of Extracted Data
data = {"Job Title": Job_Title,
       "Company Name": Company_Name,
       "Experience Required" : Experience_Required,
       "Salary Per Annum":Salary_Per_Annum}
df = pd.DataFrame(data)
df
```
